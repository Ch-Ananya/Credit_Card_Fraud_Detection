```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('creditcard.csv')
df.head()
```

```python
df.info()
df.describe()
df.isnull().sum()
```

```python
fraud_count = df['Class'].value_counts()
sns.barplot(x=fraud_count.index, y=fraud_count.values)
plt.title('Fraud vs Non-Fraud Distribution')
plt.show()
```

```python
sns.boxplot(x='Class', y='Amount', data=df)
plt.title('Transaction Amounts by Class')
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(df.corr(), cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
from sklearn.ensemble import IsolationForest
from sklearn.metrics import classification_report, confusion_matrix

# Separate features
X = df.drop('Class', axis=1)
y = df['Class']

# Fit Isolation Forest
iso = IsolationForest(contamination=0.0017, random_state=42)
y_pred = iso.fit_predict(X)

# Convert predictions: 1 â†’ non-fraud, -1 â†’ fraud
y_pred = [1 if p == -1 else 0 for p in y_pred]

print(classification_report(y, y_pred))
```

```python
sns.heatmap(confusion_matrix(y, y_pred), annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

```python
# Add prediction results
df['Predicted_Class'] = y_pred

# Aggregate by hour, amount, etc.
summary = df.groupby('Predicted_Class').agg({
    'Amount': ['count', 'mean', 'sum']
}).reset_index()

df.to_csv('fraud_results.csv', index=False)
summary.to_csv('fraud_summary.csv', index=False)
```
